## 1. 시본의 lmplot() 함수를 사용하여 다음과 같이 tip 데이터를 시각화하라. 이 데이터는 smoker 값의 Yes/No에 따라서 서로 다른 선형 회귀 직선이 하나의 그래프에 나타난다.

<br>

```python
import seaborn as sns

sns.set_theme(style="darkgrid")  # 테마 설정
tips = sns.load_dataset("tips")  # 팁 데이터 가져오기

ax = sns.lmplot(x = 'total_bill', y = 'tip', hue = 'smoker', data = tips)  # lmplot()을 이용하여 시각화
```

*시본 라이브러리를 import한다.
<br><br>
set_theme()를 이용하여 테마를 "darkgrid", 즉 어두운 격자 무늬로 설정해준다.
<br>
load_dataset()을 이용하여 tips 데이터를 가져와 tips 변수에 담는다.
<br><br>
lmplot()을 이용하여 산점도와 선형 회귀 직선을 그린다. 이때, 인자로는 x축은 total_bill, y축은 tip, hue = smoker를 이용하여 흡연자 여부 데이터를 나누어서 그려주었다.
<br><br>
이 그래프를 통해 대체적으로 총 지출 금액과 팁 액수는 비례한다는 것을 알 수 있으며, 또한 담배를 피우지 않는 사람들이 더 총 지출 금액에 비해 높은 팁을 주었다는 것을 확인할 수 있다.*

## 2. 타이타닉호는 1912년에 건조된 영국의 호화 유람선으로 건조 다시 세계 최대의 여객선이었다. 하지만 1912년의 최초이자 최후의 항해 때 빙산과 충돌해 침몰한 비운의 여객선으로 유명하다. 시본은 이 당시 타이타닉호의 생존자와 사망자 데이터를 제공하는데, 다음과 같은 방법으로 이 데이터를 읽어올 수 있다.

```python
titanic = sns.load_dataset("titanic")
```

<br><br>

### 2-1. 타이타닉 데이터를 분석하여 각 선실 등급별 남자, 여자, 어린이 데이터를 시각화하라. (countplot() 함수 이용)

<br>

```python
import matplotlib.pyplot as plt

sns.countplot(x="class", data=titanic, hue = "who")  # countplot()을 이용하여 시각화
plt.show()
```

*그래프를 그리기 위해 맷플롯립 라이브러리를 import한다.
<br><br>
countplot()을 이용하여 막대그래프를 작성한다. 이때 x 축은 클래스로 주어 사람 명 수를 카운트하고, hue = "who"로 주어 man/woman/child에 따라 세 가지 막대그래프를 그린다.
<br>
그려진 그래프를 화면에 출력한다.
<br><br>
퍼스트 클래스와 세컨드 클래스는 대체적으로 분포가 비슷한 반면, 서드 클래스가 가장 인원수가 많으며 특히 성인 남성이 많다는 것을 알 수 있다.*

<br><br>

### 2-2. 나이와 요금 사이의 관계를 알아보자. 그리고 col = 'survived' 키워드 인자를 이용하여 사망자의 분포와 생존자의 분포 간의 차이가 있는지 살펴보자. (replot() 함수 사용)

<br>

```python
sns.relplot(data=titanic, x="age", y="fare", col="survived")  # relplot()을 이용하여 산점도 그래프를 그린다.
```

*relplot()을 이용하여 산점도 그래프를 그린다.
<br>
인자로 x축은 나이, y축은 가격을 주어서 나이와 요금 사이의 관계를 알아볼 수 있도록 하였고, col = "survived"로 주어서 생존한 그룹과 사망한 그룹 사이의 분포 차이가 있는지 확인할 수 있다.
<br><br>
대체적으로 생존자들은 나이가 어리거나 요금이 높았던 경우가 많은 것을 알 수 있다.*
